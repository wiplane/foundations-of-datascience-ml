## Learning Objectives

1. Displaying random data on a scatter plot.
2. Plotting two numerical columns of a dataframe.
3. Pairplots to understand the relationship between any two variables.


```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams["figure.figsize"] = (12,8)
```

## Plotting random data on a scatter plot

```python
x = np.random.randn(50)
y = 10*np.random.randn(50)

plt.scatter(x, y)
```

## Plotting MSFT closing price vs volume of shares traded

```python
msft_df = pd.read_csv('../data/MSFT.csv', index_col=0, parse_dates=True)
msft_df.head()
```

```python
fig, ax = plt.subplots(1,1)

ax.scatter(msft_df['Close'], msft_df['Volume'])

ax.set_xlabel("Closing price")
ax.set_ylabel("Volume of Shares")

plt.show()
```
