```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
##coding the relu function

def relu_fn(x):
    return max(0, x)

vectorized_relu = np.vectorize(relu_fn)
```

## Create a few points

```python
Xs = np.arange(-10, 10)
Ys = vectorized_relu(Xs)
```

```python
print(Xs)
print(Ys)
```

## Plotting X and Y

```python
plt.plot(Xs, Ys)
```
