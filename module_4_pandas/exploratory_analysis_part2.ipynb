## Merging the two dataframes

## Cleaning Steps

1. Need to remove redundant columns
2. Cleaning or flattening the genres, cast and other columns that contain JSON data.
3. Remove duplicate rows
4. Some movies in the data have zero budget/ zero revenue, that us there valueis not recorded.
5. Change the data types of columns wherever required.
6. Replacing zero with NAN in runtime column.


```python
del_col_list = ['keywords', 'homepage', 'status', 'tagline', 'original_language', 'overview', 'production_companies',
               'original_title', 'title_y']

movies_df = movies_df.drop(del_col_list, axis=1)
movies_df.head()
```

```python
movies_df = movies_df.rename(columns={"title_x": "title"})
movies_df.head()
```

## Flattening the JSON

```python
movies_df['genres'][0]
```

```python
import json

payload = json.loads(movies_df['genres'][0])
payload
```

```python
type(payload)
```

```python
genre_list = []

for i in range(len(payload)):
    genre_list.append(payload[i]['name'])
    
genre_list
```

```python
def clean_json(x):
    genre_list = []
    for i in range(len(x)):
        genre_list.append(x[i]['name'])
    return str(genre_list)
```

```python
movies_df['genres'].apply(json.loads).apply(clean_json)
```

```python
def flatten_json(column):
    movies_df[column] = movies_df[column].apply(json.loads).apply(clean_json)
    
flatten_json('spoken_languages')
flatten_json('cast')
flatten_json('production_countries')
```

```python
flatten_json('genres')
```

```python
movies_df
```

## Dropping duplicate rows


```python
print(movies_df.shape)
movies_df = movies_df.drop_duplicates(keep='first')
print(movies_df.shape)
```

## Dropping values with zero revenue and budget

```python
movies_df.describe()
```

```python
cols = ['budget', 'revenue']

movies_df[cols] = movies_df[cols].replace(0, np.nan)
movies_df.dropna(subset=cols, inplace=True)
```

```python
movies_df.shape
```

## Change the release_date column to datetime column

```python
movies_df.info()
```

```python
movies_df["release_date"] = pd.to_datetime(movies_df['release_date'])
movies_df.info()
```

```python
movies_df['release_year'] = movies_df['release_date'].dt.year
movies_df.head()
```
