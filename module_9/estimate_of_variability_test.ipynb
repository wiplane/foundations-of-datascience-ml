```python
import pandas as pd
import numpy as np
```

```python
arr = np.array([4,3,-1,2,5])
np.var(arr)
```

```python
np.std(arr)
```

```python
range_arr = arr.max() - arr.min()
print(range_arr)
```

```python
iqr = np.percentile(arr, 75) - np.percentile(arr, 25)
print(iqr)
```
